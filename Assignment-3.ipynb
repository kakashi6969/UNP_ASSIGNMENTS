## 1)Generate 10000 values between 5 and 500. Find its mean, median, variance, 25th percentile. 

```python
x = np.random.randint(5,500,size=10000)
x
```

```python
print("mean of =",np.mean(x))
print("variance of x =",np.var(x))
print("25th percentile of x",np.percentile(x,25))
```

## 2)In the “dat.csv” replace all the null values with the mean of their respective columns (except Gender).

```python
d = pd.read_csv('dat.csv')
d
```

```python
d.isna().sum()
```

```python
for i in d:
    if i != "Gender":
        d[i].fillna(np.mean(d[i]),inplace=True)
d
```
